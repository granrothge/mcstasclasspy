```python
import mcstasclasspy.mcmonloader as mcm
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
source_shape = mcm.load_ascii_monitor('../tests/data/Source_image.dat')
```

Generate a figure the f,img and ax are all matplotlib objects that you can use the matplot lib commands on.

```python
f,ax=plt.subplots()
img =source_shape.pcolor(ax=ax)
f.colorbar(img)
ax.set_xlim((-5,5))
ax.set_ylim((-5,5))
```

Here is a command to take a 1D cut out of the above 2D data.  The syntax is cut direction, cut center, cut width

```python
cut1 = source_shape.cut('x',0,0.5)
```

Now plotting the cut with the original data

```python
f2,ax2=plt.subplots(nrows=2)
source_shape.pcolor(ax=ax2[0])
cut1.errorbar(ax=ax2[1])
for at in ax2:
    at.set_xlim((-5,5))
f2.tight_layout()
```

Now another cut

```python
cut2 = source_shape.cut('x',2,0.5)
```

plotting those cuts together

```python
f3,ax3=plt.subplots()
cut2.errorbar(ax=ax3,label=cut2.title)
cut1.errorbar(ax=ax3,label=cut1.title)
ax3.legend()
```

taking the difference

```python
cutd=cut1-cut2
```

plotting the difference

```python
f4,ax4=plt.subplots()
cutd.errorbar(ax=ax4)
```
