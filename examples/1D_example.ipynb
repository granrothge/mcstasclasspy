# Example of plotting and manipulating 1D Data.

Necessary imports

````python
import mcstasclasspy.mcmonloader as mcm
import matplotlib.pyplot as plt
%matplotlib notebook
````

Load the data for the source spectrum

````python
source = mcm.load_ascii_monitor('../tests/data/Source_spectrum.dat')
````

Most of the commands are overloaded python, numpy or matplotlib. For example the errorbar method works just like the matplotlib except you pass the axis in if you like. The tick params are then straight matplotlib

````python
f,ax =plt.subplots()
source.errorbar(ax=ax,fmt='.')
ax.tick_params(colors='magenta',which='both')
ax.yaxis.label.set_color('magenta')
````
